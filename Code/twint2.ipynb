```python
#import twint
import pandas as pd
```

```python
df = pd.read_csv("combined_csv.csv")
df["Views"] = df['Views'].map(lambda views: views.replace(",",""))
df["Views"] = pd.to_numeric(df["Views"])
print(df.sort_values("Views",ascending=False).head(30))
```

```python
new_pd = df.sort_values("Views",ascending=False).head(30)
new_pd.to_csv('TopList.csv', index = False, header=True)
```
